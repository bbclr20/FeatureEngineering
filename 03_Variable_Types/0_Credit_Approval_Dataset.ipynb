# Credit approval dataset

In this notebook, we prepare the Credit Approval data set from the UCI Machine Learning Repository.

## Download the data

To download the credit approval dataset from the UCI Machine Learning Repository:

- Visit [this website](http://archive.ics.uci.edu/ml/machine-learning-databases/credit-screening/).
- To download the data click on **crx.data**.

## Citation:

Dua, D. and Graff, C. (2019). [UCI Machine Learning Repository](http://archive.ics.uci.edu/ml). Irvine, CA: University of California, School of Information and Computer Science.

```python
import random
import pandas as pd
import numpy as np
```

```python
# Load the data
DATA_PATH = '../datasets/CreditApproval/crx.data'
data = pd.read_csv(DATA_PATH, header=None)

# Create variable names according to UCI Machine Learning information.
varnames = ['A'+str(s) for s in range(1,17)]
data.columns = varnames
```

```python
# Replace ? by np.nan.
data = data.replace('?', np.nan)

# Re-cast some variables to the correct types.
data['A2'] = data['A2'].astype('float')
data['A14'] = data['A14'].astype('float')

# Replace target values by numbers.
data['A16'] = data['A16'].map({'+':1, '-':0})
```

```python
data.head()
```

```python
# Select the random positions and set to nan
random.seed(12345)

values = list(set([random.randint(0, len(data)-1) for p in range(0, 100)]))
print(len(values))

for var in ['A3', 'A8', 'A9', 'A10']:
    data.loc[values, var] = np.nan

data.isnull().sum()
```

```python
# Categorical variables
# O for python object

cat_cols = [c for c in data.columns if data[c].dtypes=='O']
data[cat_cols].head()
```

```python
# Numerical variables
# not object

num_cols = [c for c in data.columns if data[c].dtypes!='O']
data[num_cols].head()
```
